```python
import os
from dotenv import load_dotenv
load_dotenv()

import pymongo

def get_mongodb_connection():
    mongo_url = os.getenv("MONGODB_URL")
    print("MONGO_URL:", mongo_url)  # DEBUG

    if not mongo_url:
        raise ValueError("‚ùå Missing MONGODB_URL in .env")

    client = pymongo.MongoClient(mongo_url)
    try:
        print("‚û°Ô∏è  Databases:", client.list_database_names())  # TEST CONNECTION
    except Exception as e:
        print("‚ùå MongoDB Connection Error:", e)
        raise e
    return client

def get_mongodb_collection(db_name=None, collection_name=None):
    client = get_mongodb_connection()

    db_name = db_name or os.getenv("MONGODB_DB_NAME", "VNLawsDB")
    collection_name = collection_name or os.getenv("MONGODB_COLLECTION_NAME", "VNLawsCollection")

    print("‚û°Ô∏è Using DB:", db_name)
    print("‚û°Ô∏è Using Collection:", collection_name)

    return client[db_name][collection_name]


collection = get_mongodb_collection()

print("‚û°Ô∏è Total Docs:", collection.count_documents({}))
print("‚û°Ô∏è Sample Doc:", collection.find_one())

print("‚û°Ô∏è Checking embeddings‚Ä¶")
print(collection.find_one({"embedding": {"$exists": True}}))
```
