```python
from libs.search import LegalRAGSystem

# Khởi tạo RAG system
rag = LegalRAGSystem(
    db_name="VNLawsDB",
    collection_name="VNLawsCollection",
    num_results=5
)

# Tìm kiếm với chế độ semantic
results = rag.search("Điều kiện cấp sổ đỏ", mode="semantic")
```
